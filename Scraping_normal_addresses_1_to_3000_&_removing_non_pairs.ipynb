```python
import requests
from datetime import datetime
import pandas as pd


def get_txs(address):

    starting_tx = 0

    try:
        url = 'https://blockstream.info/api/address/{}'.format(address)
        transaction_history = requests.get(url).json()
    except: 
        return('Transactions Not Found')
    
    outgoing = []
    incoming = []
    total_txs = int(transaction_history['chain_stats']['tx_count'])
    last_tx = ''

    while(starting_tx < total_txs):
        try:
            transaction_history = requests.get(url+'/txs/chain/'+last_tx).json()
        except:
            return('Transactions Not Found')
        
        for i in transaction_history:
            try:
                tx_id = i['txid']
                time = datetime.utcfromtimestamp(int(i['status']['block_time'])).strftime('%Y-%m-%d %H:%M:%S')
                inputs = []
                outputs = []
                for j in i['vin']:
                    inputs.append((j['prevout']['scriptpubkey_address'],j['prevout']['value']))
                for j in i['vout']:
                    outputs.append((j['scriptpubkey_address'],j['value']))
                
                transaction = [tx_id,time,inputs,outputs]

                isoutgoing = False
                for input_addr in transaction[2]:
                    if input_addr[0] == address:
                        isoutgoing = True
                        break
                if isoutgoing:
                    outgoing.append(transaction)
                else:
                    incoming.append(transaction)
                    
            except:
                return('Transaction data not proper')
            
        starting_tx += 50
    
    pd.DataFrame(outgoing, columns=['tx_id','time','inputs','outputs']).to_csv('transactions/' + address + '_outgoing.csv')
    pd.DataFrame(incoming, columns=['tx_id','time','inputs','outputs']).to_csv('transactions/' + address + '_incoming.csv')
        
    return('Done')
```

```python
address_list = pd.read_csv('./randomnormaladdresses.csv', index_col=0)
address_list = address_list.values.ravel()
# print(address_list)
address_list = address_list[:3000]
print(len(address_list))
```

```python
failed_addresses = []
count = 0
for address in address_list:
    # print(address)
    try:
        returned = get_txs(address)
        # print(returned)
        if returned != 'Done':
            failed = [address,returned]
            failed_addresses.append(failed)
    except:
        failed = [address,'Address not proper']
        failed_addresses.append(failed)
    count+=1
    if(count%10==0):
        print(count)

pd.DataFrame(failed_addresses, columns=['address','error']).to_csv('Failed_extraction.csv')
```
